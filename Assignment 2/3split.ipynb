```python
import os
from PIL import Image

DATA_DIR = "/content/drive/MyDrive/images/Preprocessed"
NEW_DATA_DIR = "/content/drive/MyDrive/images/Preprocessed_split3"

os.makedirs(NEW_DATA_DIR, exist_ok=True)

# For each class folder in DATA_DIR
for class_name in os.listdir(DATA_DIR):
    class_path = os.path.join(DATA_DIR, class_name)
    if not os.path.isdir(class_path):
        continue

    new_class_path = os.path.join(NEW_DATA_DIR, class_name)
    os.makedirs(new_class_path, exist_ok=True)

    # For each image in class folder
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        img = Image.open(img_path)
        width, height = img.size
        crop_width = width // 3

        # Crop and save each of the 3 parts
        for i in range(3):
            left = i * crop_width
            right = left + crop_width
            crop_img = img.crop((left, 0, right, height))

            # Save with suffix _part{i+1}
            base_name, ext = os.path.splitext(img_name)
            new_img_name = f"{base_name}_part{i+1}{ext}"
            crop_img.save(os.path.join(new_class_path, new_img_name))


```

```python

total_images = 0
for class_name in os.listdir(NEW_DATA_DIR):
    class_path = os.path.join(NEW_DATA_DIR, class_name)
    if os.path.isdir(class_path):
        # Count only files (images) in the class folder
        num_images = len([f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))])
        print(f"Class '{class_name}' has {num_images} images")
        total_images += num_images

print(f"Total images in '{NEW_DATA_DIR}': {total_images}")
```
